### Univariate Analysis
##### Analyzing single variable at a time

```python
#seperating Numerical and Categorical variables for better analysix
cat=df.select_dtypes(include=['object']).columns
num=df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables")
print(cat)
print("Numerical Variables")
print(num)
```

```python
for i in num:
    print(i)
    print('Skew :',round(df[i].skew(),2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    df[i].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sb.boxplot(x=df[i])
    plt.show()
```

```python
df
```

```python
#density plots

for i in num:
    plt.figure(figsize=(8,6))
    sb.kdeplot(df[i])
    plt.title('Density Plot for {i}')
    plt.xlabel(i)
    plt.ylabel('Density')
    plt.show()
```

```python
#violin plots 
for feature in num:
    plt.figure(figsize=(8, 6))
    sb.violinplot(x='Species', y=feature, data=df)
    plt.title(f'Violin Plot of {feature} by Species')
    plt.show()
```

```python
# Swarm Plots
for feature in num:
    plt.figure(figsize=(8, 6))
    sb.swarmplot(x='Species', y=feature, data=df)
    plt.title(f'Swarm Plot of {feature} by Species')
    plt.show()
```

### Bivariate Analysis
##### Exploring relationship between two variables 

```python
#scatter plots

pairs = [
    ('SepalLengthCm', 'SepalWidthCm'),
    ('SepalLengthCm', 'PetalLengthCm'),
    ('SepalLengthCm', 'PetalWidthCm'),
    ('SepalWidthCm', 'PetalLengthCm'),
    ('SepalWidthCm', 'PetalWidthCm'),
    ('PetalLengthCm', 'PetalWidthCm')
]


for i,j in pairs:
    plt.figure(figsize=(8,6))
    sb.scatterplot(data=df,x=i,y=j,hue="Species")
    plt.title('Scatter Plot of {i} vs {j}')
    plt.xlabel(i)
    plt.ylabel(j)
    plt.show()
```

### Correlation

##### It is a statistical phenomenon that occurs when independent variables are correlated with each other

```python
#calculating correlation matrix

correlation_matrix=df.corr()
print(correlation_matrix)
```

```python
# Visualizing matrix using HeatMap
plt.figure(figsize=(8,6))
sb.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
# Identify paris of feature with high collinearity

threshold=0.75
high_correlation=correlation_matrix[(correlation_matrix.abs()>threshold) & (correlation_matrix != 1.0)]
print(high_correlation.dropna(how="all").dropna(axis=1,how="all"))
```
